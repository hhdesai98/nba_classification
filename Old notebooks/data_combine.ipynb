```python
import pandas as pd
```

```python
b_stat = pd.read_csv('bbref_basic_stat_dat.csv')
a_stat = pd.read_csv('bbref_advanced_stat_dat.csv')
all_nba = pd.read_csv('all_nba.csv')
```

```python
b_stat = b_stat[['Player','year', 'Pos', 'Age', 'Tm', 'G', 'GS', 'MP', 'FG',
       'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT',
       'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'PTS']]
#found the data scraping error where headers were inserted into the data
# b_stat[b_stat.duplicated]
b_stat = b_stat[b_stat['Player'] != 'Player']
```

```python
a_stat = a_stat[['Player','year', 'Tm', 'PER',
       'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%',
       'TOV%', 'USG%',  'OWS', 'DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM', 'VORP']]
a_stat = a_stat[a_stat['Player'] != 'Player']
```

```python
all_stat = pd.merge(b_stat, a_stat, on = ['Player', 'year','Tm'])
all_stat
```

```python
all_stat['Player'] = all_stat['Player'].str.replace('*', '')
```

```python
all_nba = all_nba[['year',"All-NBA","Position","Player"]]
all_nba = all_nba[all_nba['year']>=1980]
all_nba
```

```python
pd.merge(all_stat,all_nba, on = ['Player', 'year'], how = "outer")
```

```python
final_df = pd.merge(all_stat,all_nba, on = ['Player', 'year'], how = "outer")
```

```python
final_df.loc[final_df['All-NBA'].isnull(),'All-NBA'] = 'None'
final_df.loc[final_df['Position'].isnull(),'Position'] = final_df.loc[final_df['Position'].isnull(),'Pos']
```

```python
final_df
```

```python
final_df.fillna(0, inplace=True)
```

```python
final_df.to_csv('final_df.csv')
```
